```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select



import time
```

```python
class Scrapping:
    def __init__(self, driver: webdriver.Chrome, index:str):
        self.driver = driver
        self.index = index
        self.wait1 = WebDriverWait(self.driver, 3)
        self.mail = f"{index}"
        self.password = f'{index}'
        self.index = index



    def Scrapping(self):

        
        self.driver.get('https://signup.live.com/signup?lcid=1033&wa=wsignin1.0&rpsnv=13&ct=1679375541&rver=7.0.6737.0&wp=MBI_SSL&wreply=https%3a%2f%2foutlook.live.com%2fowa%2f%3fnlp%3d1%26signup%3d1%26RpsCsrfState%3d43fe826b-8898-c8bb-2451-ba3b4378c656&id=292841&CBCXT=out&lw=1&fl=dob%2cflname%2cwld&cobrandid=90015&lic=1&uaid=a43b776fc5224f3281d33137341b367a')
        


        email_input = self.obtenerXpath('//*[@id="MemberName"]')
        email_input.send_keys(self.mail)
        
        
        btn_enter = self.obtenerXpath('//*[@id="iSignupAction"]')
        btn_enter.click()
        
                
        password_input = self.obtenerXpath('//*[@id="PasswordInput"]')
        password_input.send_keys(self.password)
        
        
        btn_password = self.obtenerXpath('//*[@id="iSignupAction"]')
        btn_password.click()
        
        
        name_input = self.obtenerXpath('//*[@id="FirstName"]') 
        name_input.send_keys(self.mail)
        
        lastname_input = self.obtenerXpath('//*[@id="LastName"]')
        lastname_input.send_keys(self.index)
        
    
        btn_finish = self.obtenerXpath('//*[@id="iSignupAction"]')
        btn_finish.click()
        
        
        day =  self.obtenerXpath('//*[@id="BirthDay"]')
        month = self.obtenerXpath('//*[@id="BirthMonth"]')
        year = self.obtenerXpath('//*[@id="BirthYear"]')
        
        day_select = Select(day)
        day_select.select_by_visible_text("10")

        month_select = Select(month)
        month_select.select_by_visible_text("mayo")

        year.send_keys("2000")

        btn_finish = self.obtenerXpath('//*[@id="iSignupAction"]')
        btn_finish.click()
        
        time.sleep(30)
        
        #Resolvemos el captcha antes de que ttermine la app
        self.driver.quit()



    def ObtenerListaElementos(self, xPath:str):
        return self.wait1.until(EC.presence_of_all_elements_located((
                By.XPATH, xPath)))

    def ObtenerListaElementosClassName(self, className:str):
        return self.wait1.until(EC.presence_of_all_elements_located((
                By.CLASS_NAME, className)))
    
    def obtenerXpath(self, xPath:str):
        return self.wait1.until(EC.element_to_be_clickable((
                By.XPATH, xPath)))
        
    def obtenerClassName(self, ClassName):
        return self.wait1.until(EC.presence_of_element_located((
                By.CLASS_NAME, ClassName)))




```

```python
for index in range(9,11):
    driver = webdriver.Chrome()

    web_scrapping = Scrapping(driver, str(index))

    web_scrapping.Scrapping()
```
